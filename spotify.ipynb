```python
#most popular 10 songs
top_10_songs = data.nlargest(10, 'popularity').reset_index()
top_10_songs
```

```python
# barchart of top 10 songs
# Create a bar chart to visualize the popularity of the top 10 songs of all time
plt.figure(figsize=(12, 6))
plt.barh(top_10_songs['song'], top_10_songs['popularity'], color='blue')
plt.xlabel('Popularity')
plt.ylabel('Song')
plt.title('Top 10 Songs of All Time')
plt.show()
```

```python
# Group the data by 'year' and calculate the average popularity
popularity = data.groupby('year')['popularity'].mean()
popularity
```

```python
#boxplot
plt.figure(figsize=(12, 6))
sns.boxplot(x=data['year'], y=data['popularity'])
plt.xlabel('Year')
plt.ylabel('Popularity')
plt.title('Popularity Distribution by Year')
plt.show()
```

```python
# Plot the line chart
plt.plot(popularity.index, popularity, color='green')
# Customize the plot
plt.xlabel('Year')
plt.ylabel('Average Popularity')
plt.title('Popularity by Year')

# Display the line chart
plt.show()
```

```python
# Create subplots
fig, axes = plt.subplots(nrows=3, ncols=3, figsize=(10, 7))
plt.subplots_adjust(wspace=0.3, hspace=0.5)

# Plot histograms for each attribute
attributes = ['popularity', 'danceability', 'energy', 'loudness', 'speechiness', 'acousticness', 'liveness', 'valence', 'tempo']

for i, attr in enumerate(attributes):
    row = i // 3
    col = i % 3
    
    axes[row, col].hist(data[attr], bins=20, color='skyblue', edgecolor='black')
    axes[row, col].set_title(attr.capitalize())
    axes[row, col].set_xlabel(attr.capitalize())
    axes[row, col].set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
# Bar chart of genre distribution
plt.figure(figsize=(12, 6))

genre_counts = data['genre'].value_counts()
plt.bar(genre_counts.index, genre_counts.values)
plt.xlabel('Genre')
plt.ylabel('Count')
plt.title('Genre Distribution')
plt.xticks(rotation=90)
plt.show()
```
